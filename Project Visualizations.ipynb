```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

adults = pd.read_csv('dataset/adult.data.txt', sep=", ", header=None, engine='python')
adults.columns = ['Age', 'Work_Class', 'Final_Weight', 'Education', 'Education_Num',
                  'Marital_Status', 'Occupation', 'Relationship', 'Race', 'Sex',
                  'Capital-Gain', 'Capital-Loss', 'Hours_Per_Week', 'Native_Country', 'Income']

adults = adults[adults['Work_Class'] != '?']
adults = adults[adults['Education'] != '?']
adults = adults[adults['Marital_Status'] != '?']
adults = adults[adults['Occupation'] != '?']
adults = adults[adults['Relationship'] != '?']
adults = adults[adults['Race'] != '?']
adults = adults[adults['Sex'] != '?']
adults = adults[adults['Native_Country'] != '?']
adults = adults[adults['Hours_Per_Week'] < 80]
adults = adults[adults['Hours_Per_Week'] > 20]

adults_above = adults[adults['Income'] == '>50K']
adults_below = adults[adults['Income'] == '<=50K']

fig, (ax_box1, ax_box2, ax_hist) = plt.subplots(3, sharex=True, gridspec_kw={"height_ratios": (.15, .15, .7)}, figsize=(20, 20))

sns.boxplot(adults_above['Hours_Per_Week'], ax=ax_box1, showfliers=False)
sns.boxplot(adults_below['Hours_Per_Week'], ax=ax_box2, showfliers=False)
sns.distplot(adults_above['Hours_Per_Week'], kde=False, norm_hist=True, ax=ax_hist)
sns.distplot(adults_below['Hours_Per_Week'], kde=False, norm_hist=True, ax=ax_hist)

ax_box1.set_xlabel('Hours Worked Per Week')
ax_box2.set_xlabel('Hours Worked Per Week')
ax_hist.set_xlabel('Hours Worked Per Week')
ax_box1.set_ylabel('>50K      ', rotation=0)
ax_box2.set_ylabel('<=50K      ', rotation=0)
ax_hist.legend(['>50K', '<=50K'])

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

adults = pd.read_csv('dataset/adult.data.txt', sep=", ", header=None, engine='python')
adults.columns = ['Age', 'Work_Class', 'Final_Weight', 'Education', 'Education_Num',
                  'Marital_Status', 'Occupation', 'Relationship', 'Race', 'Sex',
                  'Capital-Gain', 'Capital-Loss', 'Hours_Per_Week', 'Native_Country', 'Income']

adults = adults[adults['Work_Class'] != '?']
adults = adults[adults['Education'] != '?']
adults = adults[adults['Marital_Status'] != '?']
adults = adults[adults['Occupation'] != '?']
adults = adults[adults['Relationship'] != '?']
adults = adults[adults['Race'] != '?']
adults = adults[adults['Sex'] != '?']
adults = adults[adults['Native_Country'] != '?']
adults = adults[adults['Work_Class'] != 'Without-pay']

adults['Work_Class'] = adults['Work_Class'].replace(['State-gov', 'Federal-gov', 'Local-gov'], 'Government')
adults['Work_Class'] = adults['Work_Class'].replace(['Self-emp-inc', 'Self-emp-not-inc'], 'Self-emp')

samples = adults['Income'].value_counts()

adults_above = adults[adults['Income'] == '>50K'].sample(n=samples.min())
adults_below = adults[adults['Income'] == '<=50K'].sample(n=samples.min())

adults = pd.concat([adults_above, adults_below])

fig, axes = plt.subplots(ncols=1, nrows=1, figsize=(15, 8))
fig.subplots_adjust(hspace=.5)

mosaic(adults, ['Work_Class', 'Income'], ax=axes)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.vq import whiten

adults = pd.read_csv('dataset/adult.data.txt', sep=", ", header=None, engine='python')
adults.columns = ['Age', 'Work_Class', 'Final_Weight', 'Education', 'Education_Num',
                  'Marital_Status', 'Occupation', 'Relationship', 'Race', 'Sex',
                  'Capital-Gain', 'Capital-Loss', 'Hours_Per_Week', 'Native_Country', 'Income']

adults = adults[adults['Work_Class'] != '?']
adults = adults[adults['Education'] != '?']
adults = adults[adults['Marital_Status'] != '?']
adults = adults[adults['Occupation'] != '?']
adults = adults[adults['Relationship'] != '?']
adults = adults[adults['Race'] != '?']
adults = adults[adults['Sex'] != '?']
adults = adults[adults['Native_Country'] != '?']
adults = adults[adults['Hours_Per_Week'] < 80]
adults = adults[adults['Hours_Per_Week'] > 20]
adults = adults[adults['Age'] < 81]
adults = adults[adults['Education_Num'] > 5]

adults_above = adults[adults['Income'] == '>50K'].sample(n=50)
adults_below = adults[adults['Income'] == '<=50K'].sample(n=50)

adults = pd.concat([adults_above, adults_below])
adults = adults[['Education_Num', 'Age', 'Hours_Per_Week', 'Income']]

plot_frame = adults[['Education_Num', 'Age', 'Hours_Per_Week']]
plot_frame = whiten(plot_frame.values)
plot_frame = pd.DataFrame(plot_frame, columns=['Education_Num', 'Age', 'Hours_Per_Week'])
plot_frame['Income'] = adults['Income'].values

plt.figure(figsize=(20, 13))
pd.plotting.parallel_coordinates(plot_frame, 'Income', cols=['Education_Num', 'Age', 'Hours_Per_Week'], color=('#556270', '#C7F464'))
plt.yticks([])
plt.show()
```
